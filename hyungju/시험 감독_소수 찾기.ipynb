####13458. 시험 감독

```python
N=int(input()) #강의실 개수
A=list(map(int,input().split())) #각 강의실 당 학생수
B, C = map(int,input().split()) #총감독관, 부감독관이 각각 관리할 수 있는 학생수
a=[0 for i in range(N)] #N개의 0으로 이루어진 리스트 만들기 - 각 강의실(N개)에 필요한 감독관 수

for i in range(N):
    if A[i]-B < 1: #학생수-(총감독관이 관리할 수 있는 학생수)가 <1이면 부감독관이 필요없음
        a[i]=a[i]+1 #해당 강의실에는 총감독관만 필요
    elif A[i]-B > 0 and (A[i]-B) < C:    #학생수-(총감독관이 관리할 수 있는 학생수)가 > 0이면 부감독관이 필요, 남은 인원이 < C면 무조건 부감독 1명만 필요
        a[i]=a[i]+2
    elif A[i]-B > 0 and (A[i]-B)%2 !=0:  #학생수-(총감독관이 관리할 수 있는 학생수)가 홀수일 때는 C로 나눈 몫 + 1명의 부감독 필요
        a[i]=a[i]+(1+((A[i]-B)//C)+1)    
    else:
        a[i]=a[i]+(1+((A[i]-B)//C)) #학생수-(총감독관이 관리할 수 있는 학생수)가 짝수일 때는 C로 나눈 몫 

print(sum(a)) #각 강의실에 필요한 감독관들 수의 합
```

```python
n = int(input())
a = list(map(int, input().split()))
b, c = map(int, input().split())
result = 0

for student in a:
    student = student - b 
    result += 1 #총감독 수

    if student > 0: #남은 학생수가 있으면
        result += student // c #부감독이 맡을 수 있는 학생수로 나눠준 몫만큼 부감독 필요

        if student % c > 0: #학생수가 홀수면 나머지가 생기기 때문에 부감독 1명 추가
            result += 1

print(result)
```

####1978. 소수 찾기

```python
N=int(input()) #개수 설정
num=list(map(int,input().split())) #숫자 목록
m=[0]*N #num 개수만큼 0으로 구성된 리스트 = 각 num에 대한 약수 개수 셀 리스트
ab=0 #개수 초기값 0으로 설정

for i in range(N): 
    if num[i]==1: #num에 있는 숫자 목록 중 1이 있으면 해당 m 자리에는 0
        m[i]==0
    for j in range(2, num[i]+1): #2부터 num의 해당 숫자까지 (ex. num[i]가 3이면 2, 3)
        if num[i]%j==0: #num[i]를 1을 제외하고 2부터 num[i]로 나눴을 때 0이 나오면 = 약수
            m[i]+=1 #약수 하나 당 m 1씩 증가
    if m[i]==1: #m 리스트 중 1이 있으면 (즉 소수이면)
        ab+=1 #ab 값 1씩 증가 = 해당 num 리스트에서 소수의 개수
    
print(ab)
```
