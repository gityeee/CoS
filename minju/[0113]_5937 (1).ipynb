```python
import sys
input = sys.stdin.readline

n=int(input())

for i in range(n):
  pass = list(input().strip())
  left, right = [],[]


  if j in pass:
    if j=='<':
      if left:
        right.append(left.pop())
    elif j == '<':
      if right:
        left.append(right.pop())
    elif j == '-':
      if left:
        left.pop()
    else: 
      left.append(j)  #알파벳 삽입

    left.extend(reversed(right)) # 스택 형식(LIFO) -> 거꾸로 출력

    print(''.join(left))  # 문자열 형태 print
```
