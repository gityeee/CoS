```python
import sys
input = sys.stdin.readline

n=int(input())

for i in range(n):
  pass = list(input().strip())
  left, right = [],[]


  if j in pass:
    if j=='<':
      if left:
        right.append(left.pop())
    elif j == '<':
      if right:
        left.append(right.pop())
    elif j == '-':
      if left:
        left.pop()
    else: 
      left.append(j)  # 

    left.extend(reversed(right)) #  (LIFO) ->  

    print(''.join(left))  #   print
```
